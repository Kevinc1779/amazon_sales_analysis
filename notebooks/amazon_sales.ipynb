```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load and check
df = pd.read_csv("../data/amazon_products_sales_data_uncleaned.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum() #missing values
```

```python

# Clean rating - extract number from "4.6 out of 5 stars"
df['rating'] = df['rating'].str.extract(r'(\d+\.?\d*)').astype(float)

# Clean bought_in_last_month - extract number from "100+ bought in past month"
df['bought_in_last_month'] = df['bought_in_last_month'].str.extract(r'(\d+)').astype(float)

# Clean number_of_reviews - remove commas
df['number_of_reviews'] = df['number_of_reviews'].str.replace(',', '').astype(float)

# Clean price columns - remove $ and commas
df['listed_price'] = df['listed_price'].str.replace('$', '').str.replace(',', '').str.extract(r'(\d+\.?\d*)').astype(float)
df['current/discounted_price'] = df['current/discounted_price'].str.replace('$', '').str.replace(',', '').str.extract(r'(\d+\.?\d*)').astype(float)

# Drop columns not needed
df.drop(columns=['sustainability_badges', 'image_url', 'product_url', 'price_on_variant', 'delivery_details','collected_at'], inplace=True)

# Drop rows missing key columns
df.dropna(subset=['rating', 'bought_in_last_month'], inplace=True)

df.info()
```

```python
df
```

```python
# Convert boolean columns
df['is_best_seller'] = df['is_best_seller'].str.strip().map({'Best Seller': True, 'No Badge': False})
df['is_sponsored'] = df['is_sponsored'].str.strip().map({'Sponsored': True, 'Organic': False})
df['is_couponed'] = df['is_couponed'].apply(lambda x: False if x == 'No Coupon' else True)

# Rename messy column
df.rename(columns={'current/discounted_price': 'discounted_price'}, inplace=True)

df.head()
```

```python
# rating distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['rating'], bins=20, kde=True, color='steelblue')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# best sellers vs non best sellers by avg rating
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='is_best_seller', y='rating', color='steelblue')
plt.title('Average Rating: Best Sellers vs Non Best Sellers')
plt.xlabel('Is Best Seller')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# couponed vs non-couponed by units bought last month
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='is_couponed', y='bought_in_last_month', color='steelblue')
plt.title('Average Units Bought: Couponed vs Non-Couponed')
plt.xlabel('Has Coupon')
plt.ylabel('Avg Units Bought Last Month')
plt.tight_layout()
plt.show()
```


Companies with no coupons sold more on average the previous month. This could suggest newer or lower performaning products put out more coupon offers.

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='is_best_seller', y='bought_in_last_month', color='steelblue')
plt.title('Average Units Bought: Best Sellers vs Non Best Sellers')
plt.xlabel('Is Best Seller')
plt.ylabel('Avg Units Bought Last Month')
plt.tight_layout()
plt.show()
```

```python
df['is_best_seller'].value_counts()
```

Surprisingly, non-best-seller products averaged more monthly purchases (about 170) 
than best seller products (about 78). However, this result should be interpreted 
with caution since only 269 out of 30,773 products (less than 1%) carry the 
Best Seller badge, making the sample too small to draw reliable conclusions.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['rating'], df['bought_in_last_month'], alpha=0.3, color='steelblue')
plt.title('Rating vs Units Bought Last Month')
plt.xlabel('Rating')
plt.ylabel('Units Bought Last Month')
plt.tight_layout()
plt.show()
```


Products rated below 3.5 rarely achieve high sales volume, suggesting 
a minimum rating threshold is required to compete on Amazon. However, 
above 4.0 the relationship weakens — highly rated products show wide 
variance in purchases, indicating that rating alone does not drive sales. 
Other factors such as price, number of reviews, and category competition 
likely play a significant role beyond the rating threshold.

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='is_sponsored', y='bought_in_last_month', color='steelblue')
plt.title('Average Units Bought: Sponsored vs Organic')
plt.xlabel('Is Sponsored')
plt.ylabel('Avg Units Bought Last Month')
plt.tight_layout()
plt.show()
```

```python
df['is_sponsored'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['discounted_price'], df['bought_in_last_month'], alpha=0.3, color='steelblue')
plt.title('Price vs Units Bought Last Month')
plt.xlabel('Discounted Price ($)')
plt.ylabel('Units Bought Last Month')
plt.tight_layout()
plt.show()
```


Lower priced products dominate in sales volume.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['number_of_reviews'], df['bought_in_last_month'], alpha=0.3, color='steelblue')
plt.title('Number of Reviews vs Units Bought Last Month')
plt.xlabel('Number of Reviews')
plt.ylabel('Units Bought Last Month')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['rating', 'number_of_reviews', 'bought_in_last_month', 
                'discounted_price']].corr(), 
            annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Matrix of Key Numeric Variables')
plt.tight_layout()
plt.show()
```
