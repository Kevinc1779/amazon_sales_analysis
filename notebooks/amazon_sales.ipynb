```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load and check
df = pd.read_csv("../data/amazon_products_sales_data_uncleaned.csv")
df
```

```python

# Clean rating - extract number from "4.6 out of 5 stars"
df['rating'] = df['rating'].str.extract(r'(\d+\.?\d*)').astype(float)

# Clean number_of_reviews - remove commas
df['number_of_reviews'] = df['number_of_reviews'].str.replace(',', '').astype(float)

# Clean price columns - remove $ and commas
df['listed_price'] = df['listed_price'].str.replace('$', '').str.replace(',', '').str.extract(r'(\d+\.?\d*)').astype(float)
df['current/discounted_price'] = df['current/discounted_price'].str.replace('$', '').str.replace(',', '').str.extract(r'(\d+\.?\d*)').astype(float)

# Drop columns not needed
df.drop(columns=['sustainability_badges', 'image_url', 'product_url', 'price_on_variant', 'delivery_details','collected_at'], inplace=True)

# Drop rows missing key columns
df.dropna(subset=['rating', 'bought_in_last_month'], inplace=True)
```

```python
s = (
    df['bought_in_last_month']
    .astype(str)
    .str.upper()
    .str.replace(',', '', regex=False)
)

m = s.str.extract(r'(\d+(?:\.\d+)?)\s*([KM]?)')  # number + optional K/M
num = pd.to_numeric(m[0], errors='coerce')
mult = m[1].map({'K': 1_000, 'M': 1_000_000}).fillna(1)

df['bought_in_last_month'] = num * mult
```

```python
# Convert boolean columns
df['is_best_seller'] = df['is_best_seller'].str.strip().map({'Best Seller': 1, 'No Badge': 0}).astype('Int64')
df['is_sponsored']   = df['is_sponsored'].str.strip().map({'Sponsored': 1, 'Organic': 0}).astype('Int64')
df['is_couponed']    = df['is_couponed'].apply(lambda x: 0 if str(x).strip() == 'No Coupon' else 1).astype('Int64')
# Rename messy column
df.rename(columns={'current/discounted_price': 'discounted_price'}, inplace=True)

df.head()
```

```python
#dropping duplicates
df = df.drop_duplicates()
```

```python
# rating distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['rating'], bins=20, kde=True, color='steelblue')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# best sellers vs non best sellers by avg rating
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='is_best_seller', y='rating', color='steelblue')
plt.title('Average Rating: Best Sellers vs Non Best Sellers')
plt.xlabel('Is Best Seller')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
# couponed vs non-couponed by units bought last month
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='is_couponed', y='bought_in_last_month', color='steelblue')
plt.title('Average Units Bought: Couponed vs Non-Couponed')
plt.xlabel('Has Coupon')
plt.ylabel('Avg Units Bought Last Month')
plt.tight_layout()
plt.show()
```


Companies with no coupons sold more on average the previous month. This could suggest newer or lower performaning products put out more coupon offers.
